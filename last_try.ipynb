```python
!pip install transformers torch scikit-learn
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Or conda install -c pytorch pytorch torchvision torchaudio -c conda-forge transformers scikit-learn

model_name = "bert-base-uncased" # Or "distilbert-base-uncased", "roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3) # 3 labels: positive, negative, neutral

# Sample Data (replace with your data)
texts = ["This is great!", "This is terrible.", "This is okay."]
labels = [0, 1, 2] # 0: positive, 1: negative, 2: neutral

encoded_texts = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
train_texts, test_texts, train_labels, test_labels = train_test_split(encoded_texts["input_ids"], labels, test_size=0.2, random_state=42)

optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
model.train()
for epoch in range(3):  # Adjust number of epochs
    optimizer.zero_grad()
    outputs = model(train_texts, labels=torch.tensor(train_labels))
    loss = outputs.loss
    loss.backward()
    optimizer.step()

# Evaluation
model.eval()
with torch.no_grad():
  test_outputs = model(torch.tensor(test_texts))
  predictions = torch.argmax(test_outputs.logits, dim=-1)

# Convert predictions to a NumPy array before passing to accuracy_score
predictions = predictions.cpu().numpy()

print(classification_report(test_labels, predictions))
print("Accuracy:", accuracy_score(test_labels, predictions))
```

```python
!pip install kafka-python
from kafka import KafkaProducer, KafkaConsumer
import json

topic = "my_topic"
# Update with your active ngrok TCP address and ensure the 'tcp://' prefix is removed
brokers = "0.tcp.ngrok.io:12345"  # Removed 'tcp://'

producer = KafkaProducer(bootstrap_servers=[brokers], value_serializer=lambda v: json.dumps(v).encode('utf-8'))
consumer = KafkaConsumer(topic, bootstrap_servers=[brokers], value_deserializer=lambda v: json.loads(v.decode('utf-8')))

producer.send(topic, {'message': 'Hello from Colab!'})
producer.flush()

for message in consumer:
    print(message.value)
```
